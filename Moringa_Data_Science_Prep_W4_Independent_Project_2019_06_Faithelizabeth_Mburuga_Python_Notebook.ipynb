# Analysis

### Question 1

```
# Question 1: Identify the most popular hour of the day for picking up a shared electric
# car (Bluecar) in the city of Paris over the month of April 2018.

# To find the most popular hour of the day, I'll choose the hours between 6 and 18 since that
# is what I consider daytime

# Start by selecting the columns Bluecar counter, year, month, city and hour
#
select = electric[['Bluecar counter', 'year', 'month', 'City', 'hour']]
select
```

```
# Then select the city of Paris since that is what we are looking for
#

popular = select[select['City']  == 'Paris'] 
popular
```

```
# Select the hours between 6 and 18
#
popular_select = popular[(popular['hour']  >= 6) & (popular['hour']  <= 18)] 
popular_select
```

```
# group by hour and count the no of times so we know which is most popular
#

popular_select.groupby('hour')['hour'].count().sort_values(ascending = False)
```

```
# So the most popular hour for picking cars is 16 (4 pm) and 8 (8 am)
```

### Question 2

```
# Question 2: What is the most popular hour for returning cars?

# To get the most popular hour for returning cars, I'll make an assumption that
# the cars are returned from hour all the way to hour 5

# select the hours between that are less than 6

popular_select3 = popular[(~(popular['hour']  >= 6))]  
popular_select3
```

```
# select the hours between that are greater than 18

popular_select4 = popular[(popular['hour']  > 18)]
popular_select4
```

```
# Merge the 2 dataframes created above

popular_select5 = pd.concat((popular_select3, popular_select4), axis=0, ignore_index=True)
popular_select5
```

```
# group by hour and count the no of times so we know which is most popular
#

popular_select5.groupby('hour')['hour'].count().sort_values(ascending = False)
```

```
# So the most popular hour for returning cars is 21 (9 pm)
```
